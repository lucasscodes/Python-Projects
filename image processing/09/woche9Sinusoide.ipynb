Im Rahmen dieser Präsenzaufgabe soll eine Sinusfunktion manipuliert und so in verschiedene Sinusoide überführt werden.

```python
import matplotlib.pyplot as plt
from math import sin,cos,pi
import numpy as np
```

Definiert zunächst eine Funktion `f`, die nur den Sinus der Eingabe berechnet und plottet die Funktion an 1000 Stellen von `0` bis `10*pi`.




```python
def f(x):
  return sin(x)

x = np.linspace(0, pi*10, 1000)
y = list(map(f, x))

plt.plot(x,y)
plt.ylim((-5,5))
plt.show
```

Nun soll die Funktion `f` so verändert werden, dass der Sinus um 2 Einheiten nach oben verschoben wird. Was muss verändert werden?

```python
def f(x):
  return 2+sin(x)

x = np.linspace(0, pi*10, 1000)
y = list(map(f, x))

plt.plot(x,y)
plt.ylim((-5,5))
plt.show
```

Verändert nun die ursprüngliche Funktion `f` so, dass die Amplitude der Schwingungen dreimal so hoch ist wie vorher.

```python
def f(x):
  return 3*sin(x)

x = np.linspace(0, pi*10, 1000)
y = list(map(f, x))

plt.plot(x,y)
plt.ylim((-5,5))
plt.show
```

Verändert nun die ursprüngliche Funktion `f` so, dass die Frequenz 10-mal höher erhöht ist.

```python
def f(x):
  return sin(10*x)

x = np.linspace(0, pi*10, 1000)
y = list(map(f, x))

plt.plot(x,y)
plt.ylim((-5,5))
plt.show
```

Verändert nun die ursprüngliche Funktion `f` so, dass die Phase um `pi*0.2` verschoben wird. Was passiert, wenn die Phase um `pi` oder `pi/2` verschoben wird?

```python
def f(x):
  return sin(x+pi*.2)

x = np.linspace(0, pi*10, 1000)
y = list(map(f, x))

plt.plot(x,y)
plt.ylim((-5,5))
plt.show
```

Erzeugt nun einen Sinusoiden, der eine Phasenverschiebung von `0.2*pi`, eine Frequenz von 5 und ein verdoppelte Amplitude hat sowie um -1 auf der y-Achse verschoben ist. Wie sieht dieser aus?

```python
def f(x):
  return -1+2*sin(5*x+pi*.2)

x = np.linspace(0, pi*10, 1000)
y = list(map(f, x))

plt.plot(x,y)
plt.ylim((-5,5))
plt.show
```

Kombiniert nun zwei Sinusoiden zu einer Funktion. Der erste Sinusoid soll der normale Sinus sein, wohingegen der zweite Sinusoid ein Sinus mit einer Frequenz von 10 sein soll. Wie wird das Ergebnis aussehen?

```python
def f(x):
  return sin(x)+sin(x*10)

x = np.linspace(0, pi*10, 1000)
y = list(map(f, x))

plt.plot(x,y)
plt.ylim((-5,5))
plt.show
```
