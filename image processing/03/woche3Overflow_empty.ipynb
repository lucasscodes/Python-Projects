Zunächst importieren wir `matplotlib` und `numpy`. Außerdem laden wir das Beispielbild `page` aus `skimage.data`.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.data import page
img = page()#[:,:50]
```

So sieht das Beispielbild aus.

```python
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

Nun wird das Bild manipuliert. Es entsteht ein Überlauf, der nicht weiter behandelt wird, wenn das Bild voher einen größeren Elementtypen bekommt.

```python
imgOverflow = img.astype(np.int)*1.5
plt.imshow(imgOverflow, cmap='gray', vmin=0, vmax=255)
#Modulo bei uint8
```

Welche Art der Behandlung des Overflows würde durchgeführt werden, wenn das Bild weiterhin den Elementtypen `np.uint8` hätte?

Wie kann die Behandlung des Overflows mit "wrap around" für beliebige Elementtypen und den Wertebereich 0,...,255 implementiert werden?

```python
imgResult = imgOverflow % 256
plt.imshow(imgResult, cmap='gray', vmin=0, vmax=255)
```

Wie kann die Behandlung des Overflows mittels Setzen auf den Maximalwert implementiert werden?

```python
imgResult2 = np.minimum(imgOverflow,255)
plt.imshow(imgResult2, cmap='gray', vmin=0, vmax=255)
```
