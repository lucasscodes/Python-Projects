Schreibt eine Funktion, die zählt, wie viele Pixel einen Wert haben, der kleiner ist als der des ersten Pixels.

```python
def f1(img): #verdeckt img[x,y]
    counter = 0
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            if img[x,y]<img[0,0]:
                counter += 1
    return counter
```

Wie lässt sich das ohne `for`-Schleifen lösen?

```python
#print(len(np.where(img<img[0,0])[0]))
#print((img<img[0,0]).sum())
def f2(img):
    return np.sum(img<img[0,0])
f2(img)
```

Welche Variante ist schneller? Ruft sie dazu jeweils 100 mal auf und messt die Zeit mit time.time.


```python
from time import time

tic = time()
for i in range(100):
    f1(img)
toc = time()
print(toc-tic)

tic = time()
for i in range(100):
    f2(img)
toc = time()
print(toc-tic)
```

Wie kann man in einer Zeile prüfen, wie viele Pixel im Bereich zwischen 100 und 200 liegen?
